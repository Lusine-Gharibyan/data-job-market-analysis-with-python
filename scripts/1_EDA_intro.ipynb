## **Exploratory Data Analysis for all Data Roles**

### Roles to Explore

```python
# Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ast

# Loading Data
df = pd.read_csv("C:/Users/user/Desktop/packages/data_jobs.csv")

# Data Cleanup
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
df["job_posted_month"] = df["job_posted_date"].dt.strftime("%b")
def cleanup(skill_list):
    if pd.notna(skill_list):
        return ast.literal_eval(skill_list)
    else:
        return skill_list
df["job_skills"] = df["job_skills"].apply(cleanup)
```

```python
df_plot = df["job_title_short"].value_counts().to_frame()
sns.set_theme(style = "ticks")
sns.barplot(
    data = df_plot,
    x = "count",
    y = "job_title_short",
    legend = False,
    hue = "count",
    palette = "dark:b_r"
)
sns.despine()
plt.title("Number of Jobs per Job Title")
plt.xlabel("Number of Jobs")
plt.ylabel("")
plt.show()
```

### Countries to Explore

```python
df_plot = df["job_country"].value_counts().head(20).to_frame()
sns.set_theme(style = "ticks")
sns.barplot(
    data = df_plot,
    x = "count",
    y = "job_country",
    legend = False,
    hue = "count",
    palette = "dark:b_r"
)
sns.despine()
plt.title("Number of Jobs per Country")
plt.xlabel("Number of Jobs")
plt.ylabel("")
plt.show()
```

## **Exploratory Data Analysis for Data Analysts in the US**

### Locations to Explore:

```python
df_US = df[(df["job_country"] == "United States") & (df["job_title_short"] == "Data Analyst")].copy()
df_US_plot = df_US["job_location"].value_counts().head(10).to_frame()
sns.set_theme(style = "ticks")
sns.barplot(
    data = df_US_plot,
    x = "count",
    y = "job_location",
    legend = False,
    hue = "count",
    palette = "dark:b_r"
)
sns.despine()
plt.title("Counts of Job Locations for Data Analyst in the US")
plt.xlabel("Number of Jobs")
plt.ylabel("")
plt.show()
```

### Job Opportunities

```python
fig, ax = plt.subplots(1, 3, figsize = (15, 5))
df_US["job_work_from_home"].value_counts().plot(kind = "pie", ax = ax[0] , autopct = "%1.1f%%", startangle = 90)
ax[0].set_title("Work from Home Offered")
ax[0].set_ylabel("")
df_US["job_no_degree_mention"].value_counts().plot(kind = "pie", ax = ax[1] , autopct = "%1.1f%%", startangle = 90)
ax[1].set_title("Degree Requirement")
ax[1].set_ylabel("")
df_US["job_health_insurance"].value_counts().plot(kind = "pie", ax = ax[2] , autopct = "%1.1f%%", startangle = 90)
ax[2].set_title("Health Insurance Offered")
ax[2].set_ylabel("")
plt.show()
```

### Companies to Explore:

```python
df_US_plot = df_US["company_name"].value_counts().head(10).to_frame()
sns.set_theme(style = "ticks")
sns.barplot(
    data = df_US_plot,
    x = "count",
    y = "company_name",
    legend = False,
    hue = "count",
    palette = "dark:b_r"
)
sns.despine()
plt.title("Counts of Companies for Data Analyst in the US")
plt.xlabel("Number of Jobs")
plt.ylabel("")
plt.show()
```
